```python
import time 
import threading
import queue

def f1(x: int): 
    return x*x-x*x+4*x-x*5+x+x

def f2(x: int):
    return x+x

def calc_process(iter:int):
    result_queue = queue.Queue()
    total_result = 0

    print(f"Start process for {iter} iterations")
    def f1_process():
        for x in range(iter):
            result_queue.put(f1(x))
        
    def f2_process():
        for x in range(iter):
            result_queue.put(f2(x))
        
    def f3_process():
        nonlocal total_result
        for _ in range(2 * iter):
            total_result += result_queue.get()

    total_start = time.perf_counter()

    task_f1 = threading.Thread(target=f1_process)
    task_f2 = threading.Thread(target=f2_process)
    task_f3 = threading.Thread(target=f3_process)

    f1_start = time.perf_counter()
    task_f1.start()
    f2_start = time.perf_counter()
    task_f2.start()
    f3_start = time.perf_counter()
    task_f3.start()

    task_f1.join()
    f1_end = time.perf_counter()
    print(f"Executing time of f1() for {iter} iterations: {f1_end - f1_start:6f} seconds")
    
    task_f2.join()
    f2_end = time.perf_counter()
    print(f"Executing time of f2() for {iter} iterations: {f2_end-f2_start:6f} seconds")

    task_f3.join()
    f3_end = time.perf_counter()
    print(f"Executing time of f3() for {iter} iterations: {f3_end - f3_start:6f} seconds")
    
    total_end = time.perf_counter()
    print(f"Total time for {iter} iterations: {total_end - total_start:6f} seconds")
    print(f"Result : {total_result}")

calc_process(10_000)

calc_process(100_000)


```

```python
from multiprocessing import Process, Manager

def f1(x: int): 
    return x*x-x*x+4*x-x*5+x+x

def f2(x: int):
    return x+x

def calc_process(iter:int):
    print(f"Start process for {iter} iterations")
    
    with Manager() as manager:
        total_result = manager.Value('i', 0)

        def f1_process():
            local_sum = 0
            for x in range(iter):
                local_sum += f1(x)
            total_result.value += local_sum

        def f2_process():
            local_sum = 0
            for x in range(iter):
                local_sum += f2(x)
            total_result.value += local_sum

        total_start = time.perf_counter()

        f1_proc = Process(target=f1_process)
        f2_proc = Process(target=f2_process)
        
        f1_start = time.perf_counter()
        f1_proc.start()
        f2_start = time.perf_counter()
        f2_proc.start()

        f1_proc.join()
        f1_end = time.perf_counter()
        
        f2_proc.join()
        f2_end = time.perf_counter()

        total_end = time.perf_counter()
        
        print(f"Executing time of f1() for {iter} iterations: {f1_end - f1_start:6f} seconds")
        print(f"Executing time of f2() for {iter} iterations: {f2_end - f2_start:6f} seconds")
        print(f"Total time for {iter} iterations: {total_end - total_start:6f} seconds")
        print(f"Result : {total_result}")
        print(f"Finished process for {iter} iterations")

calc_process(10_000)

calc_process(100_000)
```
